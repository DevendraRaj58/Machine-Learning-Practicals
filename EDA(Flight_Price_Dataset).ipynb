```python
from sklearn.preprocessing import OneHotEncoder
o_encoder=OneHotEncoder()
o_encoded=o_encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
encoded_df=pd.DataFrame(o_encoded,columns=o_encoder.get_feature_names_out())
df.drop(['Airline','Source','Destination'],axis=1,inplace=True)

```

```python
df=pd.concat([df,encoded_df],axis=1)
df
```

```python
df.head()
```

```python
#Featuring date of journey
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head(2)
df.drop('Date_of_Journey',axis=1,inplace=True)
df
```

```python
df.head(2)
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Dep_Hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_Min']=df['Dep_Time'].str.split(':').str[1]
df.drop('Dep_Time',axis=1,inplace=True)
df
```

```python
df['Arr_Hour']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
df['Arr_Min']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]
df.drop('Arrival_Time',axis=1,inplace=True)
df
```

```python
df['Duration_hr']=df['Duration'].str.split(' ').str[0].str.split('h').str[0] 
df['Duration_min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df.fillna({'Duration_min': 0}, inplace=True)
```

```python
df.drop("Duration",axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import OrdinalEncoder

df['Total_Stops'].mode()
df['Total_Stops'].unique()
ord_encoder=OrdinalEncoder(categories=[['non-stop','1 stop','2 stops','3 stops','4 stops']])
df[['Total_Stops']]=ord_encoder.fit_transform(df[['Total_Stops']])
df.reset_index(drop=True,inplace=True)
```

```python
df.fillna({'Total_Stops':'1 stop'},inplace=True)
from sklearn.preprocessing import OrdinalEncoder

df['Total_Stops'].mode()
df['Total_Stops'].unique()
ord_encoder=OrdinalEncoder(categories=[['non-stop','1 stop','2 stops','3 stops','4 stops']])
df.head()
ord_encoded=ord_encoder.fit_transform(df[['Total_Stops']])
df['Total_stop']=ord_encoded
df.drop('Total_Stops',axis=1,inplace=True)
```
