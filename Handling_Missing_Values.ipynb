## Handling missing values
### Imputation

#### 1. Mean Imputation: Imputing mean value for the given category wherever null value is present 
Prefered for normally distributed data

```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\Devendra\Downloads\TitanicDataset.csv")
df.head()
```

```python
df.isna().sum()  #Age has total 177 null values
```

```python
df['Age_mean']=df['Age'].fillna(df['Age'].mean())
```

#### 2 . Median imputation : Imputing median value for the given category wherever null value is present 
Prefered for outlier prone data

```python
df['Age_median']=df['Age'].fillna(df['Age'].mode())
df
```

#### 3. Mode Imputation : Imputing mode value for the given category wherever null value is present (Preffered for categorical values)

```python
df['Embarked'].unique()  
mode_value=df[df['Embarked'].notna()]['Embarked'].mode()[0]
```

```python
print(df)    
df['Embarked']=df['Embarked'].fillna(mode_value)
```
